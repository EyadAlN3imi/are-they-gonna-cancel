```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
data = pd.read_csv('Hotel Reservations.csv')
data.head(-1)
```

**We're getting the numeric and non binary columns to calculate the corr skew and the kurtosis later**

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(data=data,y='avg_price_per_room')
plt.ylabel('avg_price_per_room'.replace('_',' ').title())
plt.subplot(1,2,2)
sns.boxplot(data=data,y='avg_price_per_room',x='booking_status')
plt.ylabel(None)
```

```python
num_col = []
for i in data.columns:
  if data[i].dtype != 'object' and len(data[i].unique()) > 2:
    num_col.append(i)
num_col
```

```python
dist = pd.DataFrame({
    "Skew": data[num_col].skew(),
    "Kurtosis": data[num_col].kurtosis()
})
dist
```

```python
fig, axs = plt.subplot_mosaic([
    ['arrival_year','booking_status'],
    ['type_of_meal_plan','market_segment_type'],
    ['room_type_reserved','room_type_reserved']],layout='constrained',figsize=(12,7))

dc = {
    'arrival_year': 'Set1',
    'type_of_meal_plan': 'Set2',
    'booking_status': 'Set3',
    'market_segment_type': 'viridis',
    'room_type_reserved': None
}

for k,v in dc.items():
  pl = sns.color_palette(v,n_colors=data[k].nunique())
  counts = data[k].value_counts()
  axs[k].set_title(k.replace('_',' ').title())
  if k == 'arrival_year':
        bars = axs[k].bar(counts.index.astype(str), counts.values, color=pl)
  else:
        bars = axs[k].bar(counts.index, counts.values, color=pl)

  for bar in bars:
        height = bar.get_height()
        axs[k].annotate(f'{height}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom')

  axs[k].spines['top'].set_visible(False)
  axs[k].spines['right'].set_visible(False)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.histplot(data=data,x='avg_price_per_room',bins=50)
plt.xlabel('avg_price_per_room'.replace('_',' ').title(),size=12)
plt.subplot(1,2,2)
sns.histplot(data=data,x='avg_price_per_room',bins=50,hue='booking_status')
plt.xlabel('avg_price_per_room'.replace('_',' ').title(),size=12)
plt.ylabel(None)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data, x='lead_time', y='avg_price_per_room', hue='booking_status')
```

```python
corr = data[num_col].corr()
plt.figure(figsize=(13,4))
heatmap = sns.heatmap(corr, annot=True,mask=np.triu(corr), annot_kws={'fontsize': 10})
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=30, ha='right')
plt.show()
```

```python
crosstab_result = pd.crosstab(data['booking_status'], data['arrival_month'])

plt.figure(figsize=(12, 5))
sns.heatmap(crosstab_result, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Booking Status vs Arrival Month')
plt.xlabel('Arrival Month')
plt.ylabel('Booking Status')

plt.show()
```
